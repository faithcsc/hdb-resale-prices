# Fetch data

```python
original_data_filename = "hdb-resale-prices.csv"
today_date = datetime.now().date()
data_not_older_than_X_days = 7

def download_records():
    print("Downloading records...")
    base_url = "https://data.gov.sg"
    url = "/api/action/datastore_search?resource_id=f1765b54-a209-4718-8d38-a39237f502b3&limit=10000000"
    res = []
    while True:
        r = requests.get(base_url + url)
        r = json.loads(r.content.decode("utf-8"))
        records = r.get("result", {}).get("records", {})
        if not records:
            print("Downloaded")
            return res
        res.extend(records)
        url = r["result"]["_links"]["next"]

def save_records(records, date, filename):
    print("Saving records dated ", date, "...", sep="")
    item = {
        "cache_date": date,
        "records": records
    }
    with open(filename, "w") as f:
        f.write(json.dumps(item))
    print("Saved")

def read_records(filename):
    print("Reading records from file...")
    try:
        with open(filename) as f:
            data = json.loads(f.read())
        cache_date = datetime.strptime(data["cache_date"], "%Y-%m-%d").date()
        if today_date - cache_date < timedelta(days=data_not_older_than_X_days):
            print("Read records, cache date:", cache_date)
            return data["records"]
        print("Cached data from ", cache_date, ", too old", sep="")
    except Exception as e:
        print("Could not read records: ", e)
    return {}
        
def get_records():
    records = read_records(original_data_filename)
    if not records:
        records = download_records()
        save_records(records, today_date.strftime("%Y-%m-%d"), original_data_filename)
    return records
        
records = get_records()
```

# Clean data

```python
def nearest_int(original):
    return int(original + 0.5)

def bin_5(original):
    # Eg. 40..44 -> 40, 45..49 -> 45
    original = int(float(original))
    remainder = original % 10
    quotient = original - remainder
    if remainder >= 5:
        quotient += 5
    return quotient

def bin_10(original):
    # Eg. 40..49 -> 40
    original = int(float(original))
    remainder = original % 10
    quotient = original - remainder
    return quotient

def parse_month(original):
    return datetime.strptime(original, "%Y-%m").date()

def parse_lease(original):
    years = int(original.split(" ")[0])
    months = 0
    if "month" in original:
        months = int(original.split(" ")[2])
    return round(float(years) + (float(months)/12), 1)

def parse_storey(original):
    return int(original.split(" ")[0])

keep_cols = ["month", "town", "street_name", "block", "flat_type", "floor_area_sqm", "storey_range", "remaining_lease", "resale_price"]

df = pd.DataFrame(records)
df["town"] = df["town"].apply(lambda x: x.title())
df["month"] = df["month"].apply(lambda x: parse_month(x))
df["storey_range"] = df["storey_range"].apply(lambda x: parse_storey(x))
df["resale_price"] = df["resale_price"].apply(lambda x: int(float(x)/1000))
df["remaining_lease"] = df["remaining_lease"].apply(lambda x: parse_lease(x))
df["floor_area_sqm"] = df["floor_area_sqm"].apply(lambda x: int(float(x)))
df = df[keep_cols]
df = df.sort_values(by=keep_cols)
df
```

# Visualize data

```python
def force_type(val):
    try:
        return int(val)
    except:
        return val 

def plot_hist_with_bin_width(df_col, binwidth):
    min_val = int(min(df_col.astype(float)))
    max_val = int(max(df_col.astype(float)))
    plt.hist(df_col, bins=range(min_val, max_val + binwidth, binwidth))
    plt.show()

cols = ["flat_type", "floor_area_sqm", "storey_range", "remaining_lease", "resale_price"]

for col in cols:
    print("#" * 50)
    print(col)
    if type(force_type(df[col][0])) == int:
        plot_hist_with_bin_width(df[col], 5)
    else:
        print(df[col].value_counts())
```

# Filter data

```python
filters = {
    "town": lambda x: True,
    "floor_area_sqm": lambda x: x >= 70,
    "remaining_lease": lambda x: x >= 80,
    "month": lambda x: today_date - timedelta(days=365) <= x <= today_date,
    "storey_range": lambda x: x >= 4,
    "street_name": lambda x: True,
    "block": lambda x: True,
    "flat_type": lambda x: True,
}
group_by = [
    "town",
    "floor_area_sqm",
    "remaining_lease",
]

valid_transactions = []
for i, row in df.iterrows():
    for var_name, var_fn in filters.items():
        if not var_fn(row[var_name]):
            break
    else:
        valid_transactions.append(i)
df_valid = df.loc[valid_transactions]
df_valid["floor_area_sqm"] = df_valid["floor_area_sqm"].apply(lambda x: bin_10(x))
df_valid["remaining_lease"] = df_valid["remaining_lease"].apply(lambda x: bin_5(x))

results = df_valid.groupby(group_by)["resale_price"].agg(['min', 'max', 'mean', 'count']).reset_index()
results = results.sort_values(by=["mean"], ignore_index=True)
for col in results.columns:
    if results[col].dtype.kind in "iufc":
        results[col] = results[col].astype(int)
pd.set_option("display.max_rows", len(results))
results
```
